```python
import pandas as pd
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
```

```python
url='https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user'
data=pd.read_csv(url, sep='|')
data.head()
```

```python
mean_age_per_occupation = data.groupby('occupation')['age'].mean()
print(mean_age_per_occupation)
```

```python
pd.DataFrame(data = {'min':data.groupby(['occupation']).min()['age'],'max':data.groupby(['occupation']).max()['age']})
```

```python
mean_age_per_occupation_gender = data.groupby(['occupation', 'gender'])['age'].mean()

print(mean_age_per_occupation_gender)
```
